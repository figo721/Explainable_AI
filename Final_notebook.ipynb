## Model and Dataset

#### Model

The purpose of this experiment is to predict students' final math performance in secondary schools and investigate the potential effects of various variables on the result. We employ the multilayer perceptron regression model from the Scikit-Learn Python library to carry out this experiment. This model lacks weight initialization and an activation function in the output layer, among other things. The squared error function serves as this model's loss function as well. In this model, three hidden layers with 64 neurons each are used for simplicity. The MLP regression model thus has five layers, including input and output layers. The weight optimizer employs stochastic gradient descent, and the RELU function serves as each neuron's activation function for each hidden layer.

- Predict students' final math performance in secondary schools 
- Investigate the potential effects of various variables on the result
- MLP Regression Model with three hidden layers and 64 neurons each hidden layer
- Stochastic gradient descent, and the RELU function 


#### Dataset

The Student Performance Data Set, which can be found at https://archive.ics.uci.edu/ml/datasets/student+performance, is the dataset used for this experiment. The data was gathered 395 students from two secondary schools in Portugal using surveys and school reports. The dataset offers insightful data to investigate how the attributes affect the student's mathematics performance. The dataset includes data that may have an indirect effect in addition to information that appears to have a direct impact on the subject's final score. As a result, I think this dataset is appropriate for studying how the final score predicted based on the information from chosen qualities was actually attained.

```python
# Please note that the code is mainly from the following site:
# https://www.justintodata.com/explainable-machine-learning-with-python/
# use different train model 

import pandas as pd
import shap
import sklearn

# read the dataset
df = pd.read_csv('student-mat.csv', delimiter=';')
# drop columns that are less related to the target 
cols_to_drop = ['sex','internet','nursery','Pstatus','famsize','school', 'age', 'address', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'famsup', 
                'romantic', 'goout', 'Dalc', 'Walc', 'health', 'G1', 'G2']
# rename the columns so that they are understandable
df = df.drop(columns=cols_to_drop).rename(columns={  
                                                   'traveltime': 'travel_time_school', 'studytime': 'study_time', 
                                                   'failures': 'past_failures', 'schoolsup': 'extra_support', 
                                                   'paid': 'extra_paid_classes', 'activities': 'extra_curricular', 
                                                   'higher': 'want_higher_edu', 'famrel': 'family_relationships', 
                                                   'freetime': 'free_time', 'absences': 'number_absences', 
                                                   'G3': 'score'})
# convert categorical variables into dummy variables
df = pd.get_dummies(df, drop_first=True)
# define the features
df_features = df.drop(columns='score')
df_features.info()

reg = sklearn.neural_network.MLPRegressor(hidden_layer_sizes = (64,64,64),activation = 'relu', solver = 'sgd',learning_rate = 'constant', random_state=1, max_iter=2000).fit(df_features, df['score'])
```

## Partial Dependence Plots (PDP)

One of several model-independent and summary-based approaches for displaying the marginal effect of one or two attributes in a dataset is the partial dependence plot (PDP). This method's goal is to provide the global behavior of a trained model and to comprehend the relationship between the predictÂ target and the feature (s). The major limitation of this model explanation is that an assumption of this approach is that a selected feature on the plot has no correlation with the remaining features.

#### Why
Interpretability and Explainability - In terms of its capacity to in-depth research and examine a deep learning model, this explanation technique has limits. The general pattern or behavior of a trained deep learning model can still be understood using this way. It is also an effective technique for researching how particular features may influence a deep learning model's prediction.

Model debugging and improvement - This technique can be used to enhance functionality and reduce the dimensionality of input data as The relationship between a chosen feature and the chosen decision can be revealed using the partial dependency plot method in deep learning.

Comparing and Choosing Models - This approach is not appropriate for either comparing or choosing models because its goal is to explore how particular attributes affect a trained model's choice.

Teaching Deep Learning Concepts - PDP is not the best way for giving information to someone who already knows about AI and ML to update their knowledge of a deep learning model. However, using this tool would help people grasp the fundamental, non-technical idea of AI and ML.

#### Who

Model Developers and Builders - The summary-based method is ineffective for anyone attempting to develop or enhance a complicated model since it does not provide meaningful and specific information in deep learning models and model training.

Model Users - PDP is an option for those looking for a straightforward explanation of a deep learning model for a simple task.

Non-experts - This explanation technique is simple enough for non-expert users to understand and covers the fundamental knowledge needed to comprehend the idea of deep learning models.

#### What

PDP is used to comprehend a deep learning model's overall behavior rather than its specifics. Therefore, with the exception of the criterion for aggregated information, this technique is inappropriate for this area. Although this approach doesn't offer instance-level activations, it can be used to examine the connection between a dataset's features and a deep learning model's final prediction.

#### How

Node-link Diagrames for Network Architectures - Since this method is summary-based, we are unable to understand the dataflow and model architecture of neural networks using it.

Dimensionality Reduction and Scatter plots - This technique is inappropriate for this criterion because this model does not offer instance-level activations.

Line Charts for Temporal Metrics - A summary-based method is not appropriate for tracking the progress of model training for each epoch.

Instance-based Analysis and Exploration - Because this is a summary-based technique, it is not optional for investigating and explaining specific data instances in a deep learning model.

Interactive Experimentation - The deep learning model cannot be explored and investigated through an interactive interface using this method.

Algorithms for Attribution & Feature Visualization - The PDP approach may be a viable choice to investigate and comprehend which characteristics or aspects are more crucial or significant to a deep learning model's decision.

#### When

This method can only investigate a deep learning model after the model's training is complete.

#### Where

Institute of Mathematical Statistics at JSTOR - http://www.jstor.org/stable/2699986 .

## Shapley Additive Explanations (SHAP) 


A machine learning model's output can be shown using the visualization tool SHAPley Additive Explanations to make it easier to understand. By calculating how much each feature contributed to the prediction, it can be used to explain the result of any model. The Shapely values from coalitional game theory serve as the foundation for this instance-based explanation method. SHAPley values are a key component of the SHAP tool because they allow SHAP to connect optimal credit allocation with local explanations. When we talk about SHAPley values, we can think of them as a method for determining how to distribute the contribution of the features among the features as accurately as possible.

#### Why

Interpretability and Explainability - SHAP is a collection of global and local methods that allow us to provide hints to understand the overall structure of how a model makes a decision as well as the model's decisions for a single instance. We could understand the significance of each feature or attribute of a dataset in a deep learning model based on the Shapley values.

Debugging and Improving models - This technique is useful for improving performance and reducing the dimensionality of input data because it provides both summary and instance interpretability information to understand deep learning models.

Comparing and Selecting models - This method cannot be used to compare or choose models because its goal is to explore how particular attributes affect a trained model's final decision.

Teaching Deep Learning Concepts - This interpretable method, which provides both global and local information, allows one to understand the final prediction of a deep learning model. Due to the nature of the technique, it provides information to grasp a deep learning model's basic idea for managing an easy problem as well as to reinforce the memory of a deep learning model for individuals who are already familiar with AI or ML.

#### Who
Model Developers and Builders - This method is an optional tool for individuals who wish to build a deep learning model for a problem that isn't extremely complex because it gives information in both global and local contexts. Because it is unable to look at specific model structures and parameters, this technique is not the best tool for solving extremely complex problems.

Model Users - SHAP is one of ideal techniques for model users who want to solve various ML problems.

Non-experts - This way to explanation provides fundamental knowledge to comprehend the idea of deep learning models and is usable by non-expert users.

#### What 
Although this method is for both global and local explainability, it does not offer comprehensive data to research the deep learning model's architecture. Therefore, all criteria fail to satisfy the criteria, except forÂ the criteria of aggregated information in this section. The SHAP approach enables the extraction of global behavior information from a deep learning model by providing instance-based interpretable information in aggregated data.

#### How

Node-link Diagrames for Network Architectures - The dataflow and model architecture of neural networks are not covered by this explainable method. Therefore, this method does not meet this requirement.

Dimensionality Reduction and Scatter plots - SHAP does not provide an embedding method or scatter plot technique to investigate a deep learning model.

Line Charts for Temporal Metrics - This method cannot be used to track the progress of model training at each epoch since it can study a deep learning model after training the model.

Instance-based Analysis and Exploration - Because it offers quite in-depth details on the relative importance of each instance for prediction, SHAP may be a viable option for comprehending the deep learning model by instance scenario.

Interactive Experimentation - This explainable tool has interactive experimentation functionality that enables us to look at the impact of various inputs, even though we haven't used interactive functionality in this experiment.

Algorithms for Attribution & Feature Visualization - Because this technique provides a ranking of feature importance for both the overall model and each instance of the model, it is very useful information for determining which features are more influential to the model's final prediction.

This method can only investigate a deep learning model after the model's training is complete.

```python
# Global Explainer
explainer = shap.Explainer(reg.predict, df_features)
shap_values = explainer(df_features)
shap.summary_plot(shap_values, df_features)
```

The above plot depicts the overall structure of how the model predicts a mathematics subject's final score. The plot shows that the "number of absences" feature is the most important factor in predicting the final score of the mathematics subject. And the "free time" feature is the model's third important feature, and the red color dots (higher values) have a positive impact on the model's final prediction.



```python
# Local Explainer
# 10th student example
shap.plots.waterfall(shap_values[10])

# 100th student example
shap.plots.waterfall(shap_values[100])
```

The two plots above are examples of the instance-based (local) explanation method. As can be seen, each instance (student) yields a different result in order to arrive at the final prediction. In the first plot, the student has an expected final score of 10.628 in the mathematics subject. The most important feature for student prediction is "number of absences," which has a negative impact on the final mathematics score. The student's "past failures" are the second most important factor influencing the final score.

On the other hand, the second plot differs from the first plot of the instance-based explanation method in this experiment. The most important feature of the second plot student, in contrast to the first, is "extra support yes," which indicates that the student receives additional educational support, which has a negative effect on his or her predicted score. In addition, the final expected score for the mathematics subject is 8.761. The third significant feature of the second plot example student is "free time," which has a positive effect on the student's expected final mathematics score.

The importance of features varies depending on the instance, as shown in the last two plots. As a result, the SHAP method is used not only to obtain overall interpretable information about the model, but also for each instance, which is important for understanding a deep learning model.
